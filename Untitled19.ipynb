```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
import cv2 as cv2
filename = ('/content/drive/MyDrive/PCVK/Brain Tumor/Y8.jpg')
img = cv2.imread(filename,0) 
blur = cv2. GaussianBlur(img, (5,5),0)

ret, th1 = cv2.threshold(blur, 160, 255, cv2.THRESH_BINARY)
 
contours, hierarchy = cv.findContours(th1, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)

cv.drawContours(img, contours, -1, (0,0,255), 5)

titles = ['Citra Asli', 'Hasil global threshold', 'Hasil Deteksi tumor'] 
citra = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
citra3 = [blur, th1, citra]

plt. figure(figsize = (15,15)) 
for i in range(len(citra3)):
    plt. subplot(1,3,i+1),plt.imshow(citra3[i], 'gray') 
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([]) 
plt.show()
```
